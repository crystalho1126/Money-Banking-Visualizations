```python
# Import the pandas library as pd
import pandas as pd

# Import the plotting library matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

# Set plotting style to 'classic' because I like the look
plt.style.use('classic')

# Specify that you want plots displayed in the Jupyter Notebook and not in an external window (IPython magic command)
%matplotlib inline

# Download the M1 & M2 data
df = pd.read_excel(r'C:\Users\Crystal Ho\Documents\Spring 2019 - Grad\Econ 161A Money and Banking\Econ_161A_M1_M2.xls', 'Monthly', sep='\s+',index_col=0, parse_dates = True)

# Print first 5 rows of M1M2 variable
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(df['M1SL'], alpha=0.75, label='M1 Money Stock')
ax.plot(df['M2SL'], alpha=0.75, label='M2 Money Stock')
ax.legend(loc='upper left')
ax.set_xlabel('Year')
ax.set_ylabel('Billions of Dollars')
ax.grid()
ax.set_title('FRED M1 M2 Money Stock (1959-2015)')
```

```python
# Download the Fed Funds Rate data
df = pd.read_excel(r'C:\Users\Crystal Ho\Documents\Spring 2019 - Grad\Econ 161A Money and Banking\Federal_Funds_Rate_xls\Federal_Funds_Rate.xls','Daily,_7-Day', index_col=0, parse_dates = True)
df = df.dropna()
# Print first 5 rows of Funds Rate variable
df.head()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.plot(df['DFF'], alpha=0.75, label= "Effective Fed Funds Rate")
ax.plot(df['DFEDTAR'], alpha=0.75, label= "Fed Funds Target Rate")

ax.legend(loc='upper right')
ax.set_ylim(0,8)
ax.set_xlabel('Date')
ax.set_ylabel('Percent')
ax.grid()
ax.set_title('FRED Federal Funds Rates (2000-2008)')
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.plot(df['DFEDTAR']-df['DFF'], alpha=0.75, label= "Effective Fed Funds Rate Minus Target Rate")

ax.legend(loc='upper right')
ax.set_ylim(0,2)
ax.set_xlabel('Date')
ax.set_ylabel('Percent')
ax.grid()
ax.set_title('FRED Federal Funds Rate Difference From Target (2000-2008)')
```
